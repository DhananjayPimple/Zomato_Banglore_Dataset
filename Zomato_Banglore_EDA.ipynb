```python
# Importing all the necessaries libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Removing all the columns which are not considered for analysis

zom_df = zom_df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis=1)
zom_df.head()
```

```python
zom_df.info()
```

## Cleaning The Data

```python
# Droping the duplicate rows
zom_df.drop_duplicates(inplace = True)
zom_df.shape
```

```python
zom_df['rate'].unique()
```

```python
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]               # value at 0th index i.e. 4.1/5 >> ['4.1','5']
        return float(value)

zom_df['rate'] = zom_df['rate'].apply(handlerate)
zom_df.head()
```

```python
zom_df.rate.isnull().sum()
```

```python
# Filling null values in 'rate' column with mean value

zom_df['rate'].fillna(zom_df['rate'].mean(), inplace=True)
zom_df.rate.isnull().sum()
```

```python
zom_df.info()
```

```python
# Droping all the null values (which are very few ) in columns such as 'location','rest_type','cuisines','approx_cost(for two people)' etc

zom_df.dropna(inplace=True)
zom_df.head()
```

```python
# Renaming columns having lengthy names 
zom_df.rename(columns = {'approx_cost(for two people)':'cost2people', 'listed_in(type)':'type','listed_in(city)':'city'}, inplace=True)
zom_df.head()
```

```python
zom_df['location'].unique()
```

```python
zom_df['city'].unique()
```

```python
zom_df = zom_df.drop(['city'], axis=1)
```

```python
zom_df['cost2people'].unique()
```

```python
# Removing ',' from the 'cost2people' column

def handlecomma(value):
    value = str(value)
    if ',' in value :
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
zom_df['cost2people'] = zom_df['cost2people'].apply(handlecomma)
zom_df['cost2people'].unique()
```

```python
zom_df.head()
```

```python
zom_df['rest_type'].value_counts()
```

```python
rest_types = zom_df['rest_type'].value_counts()
rest_types
```

```python
rest_type_lessThan1000 = rest_types[rest_types<1000]
rest_type_lessThan1000
```

```python
def handle_rest_type(value):
    if (value in rest_type_lessThan1000):
        return 'others'
    else:
        return value
    
zom_df['rest_type'] = zom_df['rest_type'].apply(handle_rest_type)
zom_df['rest_type'].value_counts()
```

```python
zom_df['location'].value_counts()
```

```python
location = zom_df['location'].value_counts(ascending=False)

location_lessThan300 = location[location<300]

def handle_location(value):
    if (value in location_lessThan300):
        return 'others'
    else:
        return value

zom_df['location'] = zom_df['location'].apply(handle_location)
zom_df['location'].value_counts()
```

```python
zom_df.head()
```

```python
zom_df['cuisines'].value_counts()
```

```python
cuisines = zom_df['cuisines'].value_counts()

cuisines_lessThan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if (value in cuisines_lessThan100):
        return 'others'
    else:
        return value
    
zom_df['cuisines'] = zom_df['cuisines'].apply(handle_cuisines)
zom_df['cuisines'].value_counts()    
```

```python
zom_df['type'].value_counts()
```

## 2) Find the top 5 most and least favorite cuisines ?

```python
cuisines_df = zom_df.groupby('cuisines')['votes'].sum()
cuisines_df
```

```python
top_six = cuisines_df.sort_values(ascending = False).head(6)
top_six
```

```python
# visualization

plt.figure(figsize = (10, 6))
plt.title('Most favourite cuisines')
sns.barplot(x = top_five.index, y = top_five, palette = 'Pastel1')
plt.xticks(rotation = 90)
```
